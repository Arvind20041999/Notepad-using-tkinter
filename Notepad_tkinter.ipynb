```python
import tkinter
import os
from tkinter import *
```

```python
# To get the space above for message 
from tkinter.messagebox import *
```

```python
# To get the dialog box to open when required 
from tkinter.filedialog import *
```

```python
class Notepad:
    root=Tk()
    
    #default window width and height
    thiswidth=300
    thisheight=300
    textarea=Text(root)
    menubar=Menu(root)
    filemenu=Menu(menubar,tearoff=0)
    editmenu=Menu(menubar,tearoff=0)
    helpmenu=Menu(menubar,tearoff=0)
    
    #Adding scrollbar
    scrollbar=Scrollbar(textarea)
    file=None
    def __init__(self,**kwargs):
        #set icon
        try:
            self.root.wm_iconbitmap("Notepad.ico")
        except:
            pass
        #set window size
        try:
            self.thiswidth=kwargs['width']
        except KeyError:
            pass
        try:
            self.thisheight=kwargs['height']
        except KeyError:
            pass
        
        #set window title
        self.root.title("Notepad Editor")
        
        #center the window
        screenwidth=self.root.winfo_screenwidth()
        screenheight=self.root.winfo_screenheight()
        
        #left alling
        left=(screenwidth/2)-(self.thiswidth/2)
        #right alling
        top=(screenheight/2)-(self.thisheight/2)
        
        #For top and bottom 
        self.root.geometry('%dx%d+%d+%d' % (self.thiswidth,self.thisheight,left,top))
        
        #to make text area auto resizeable
        self.root.grid_rowconfigure(0,weight=1)
        self.root.grid_columnconfigure(0,weight=1)
        
        #adding controls
        self.textarea.grid(sticky=N+E+S+W)
        
        #for opening new file
        self.filemenu.add_command(label="New",command=self.newFile)
        #for opening existing file
        self.filemenu.add_command(label="Open",command=self.openFile)
        #for saving current file
        self.filemenu.add_command(label="Save",command=self.saveFile)
        self.filemenu.add_separator()
        self.filemenu.add_command(label="Exit",command=self.quitApplication)
        self.menubar.add_cascade(label="File",menu=self.filemenu)
        
        #adding feature of cut copy and paste
        self.editmenu.add_command(label="Cut",command=self.cut)
        self.editmenu.add_command(label="Copy",command=self.copy)
        self.editmenu.add_command(label="Paste",command=self.paste)
        self.menubar.add_cascade(label="Edit",menu=self.editmenu)
        
        #adding description feature
        self.helpmenu.add_command(label="About",command=self.showAbout)
        self.menubar.add_cascade(label="Help",menu=self.helpmenu)
        
        #Adding self adjustment features
        self.root.config(menu=self.menubar)
        self.scrollbar.pack(side=RIGHT,fill=Y)
        self.scrollbar.config(command=self.textarea.yview)
        self.textarea.config(yscrollcommand=self.scrollbar.set)
     
    #exit notepad
    def quitApplication(self):
        self.root.destroy()
    def showAbout(self):
        showinfo("Notepad Created by Arvind")
    def openFile(self):
        self.file=askopenfilename(defaultextension=".txt",filetypes=[("AllFiles","*.*"),("TextDocuments","*.txt")])
        if(self.file==""):
            self.file=None
        else:
            self.root.title(os.path.basename(self.file)+" -Notepad ")
            self.textarea.delete(1.0,END)
            file=open(self.file,"r")
            self.textarea.insert(1.0,file.read())
            file.close()
    def newFile(self):
        self.root.title("Untitled_Notepad")
        self.file=None
        self.textarea.delete(1.0,END)
    def saveFile(self):
        if(self.file==None):
            self.file=asksaveasfilename(initialfile="untitled.txt",defaultextension=".txt",filetypes=[("AllFiles","*.*"),("TextDocument","*.txt")])
            if(self.file==""):
                self.file=None
            else:
                file=open(self.file,'w')
                file.write(self.textarea.get(1.0,END))
                file.close()
                self.root.title(os.path.basename('self.file')+" -Notepad ")
        else:
            file=open(self.file,'w')
            file.write(self.textarea.get(1.0,END))
            file.close()
            
    def cut(self):
        self.textarea.event_generate("<<Cut>>")
    def copy(self):
        self.textarea.event_generate("<<Copy>>")
    def paste(self):
        self.textarea.event_generate("<<Paste>>")
    def run(self):
        self.root.mainloop()
            
notepad=Notepad(width=600,height=400)
notepad.run()
```
